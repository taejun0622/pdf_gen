```python


import os
import requests
from pathlib import Path
from reportlab.lib.units import mm
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
from PIL import Image, ImageDraw
from reportlab.lib.units import mm
from pathlib import Path
from reportlab.lib.colors import Color  


def create_b5_canvas_with_outer_edge_crop_marks(c, page_width, page_height):
    # 제단선 설정
    margin = 5 * mm  # PDF 끝에서 B5 영역까지의 거리 (여백)
    bleed = 3 * mm  # PDF 끝에서 제단선 시작 위치까지 거리 (총 8mm)
    trim_mark_length = 5 * mm  # 제단선 길이
    trim_mark_thickness = 0.25  # 제단선 두께
    c.setLineWidth(trim_mark_thickness)

    # 각 모서리에서 8mm 떨어진 위치에서 시작하여 바깥쪽으로 5mm 길이의 제단선을 그리기
    # 왼쪽 아래 모서리
    c.line(margin + bleed, 0, margin + bleed, trim_mark_length)  # 수직선 (위로 5mm 뻗음)
    c.line(0, margin + bleed, trim_mark_length, margin + bleed)  # 수평선 (오른쪽으로 5mm 뻗음)

    # 왼쪽 위 모서리
    c.line(margin + bleed, page_height, margin + bleed, page_height - trim_mark_length)  # 수직선 (아래로 5mm 뻗음)
    c.line(0, page_height - margin - bleed, trim_mark_length, page_height - margin - bleed)  # 수평선 (오른쪽으로 5mm 뻗음)

    # 오른쪽 아래 모서리
    c.line(page_width - margin - bleed, 0, page_width - margin - bleed, trim_mark_length)  # 수직선 (위로 5mm 뻗음)
    c.line(page_width, margin + bleed, page_width - trim_mark_length, margin + bleed)  # 수평선 (왼쪽으로 5mm 뻗음)

    # 오른쪽 위 모서리
    c.line(page_width - margin - bleed, page_height, page_width - margin - bleed, page_height - trim_mark_length)  # 수직선 (아래로 5mm 뻗음)
    c.line(page_width, page_height - margin - bleed, page_width - trim_mark_length, page_height - margin - bleed)  # 수평선 (왼쪽으로 5mm 뻗음)

    return c

def download_file(url, save_path):
    try:
        response = requests.get(url, stream=True)
        response.raise_for_status()  # 요청이 실패하면 예외 발생
        with open(save_path, "wb") as file:
            for chunk in response.iter_content(1024):
                file.write(chunk)
        print(f"Downloaded: {save_path}")
    except requests.exceptions.RequestException as e:
        print(f"Failed to download {url}: {e}")

def process_magazine_data(magazine_data=dict, call_time=str):
    # 호출 시각을 기준으로 폴더 생성
    base_dir = Path(f"temp/{call_time}")
    base_dir.mkdir(parents=True, exist_ok=True)

    # 참가자 프로필 이미지 다운로드
    participants_dir = base_dir / "participants"
    participants_dir.mkdir(exist_ok=True)
    for i, participant in enumerate(magazine_data["participants"], start=1):
        file_name = f"participant_{i}_{participant['member_id']}.jpg"
        save_path = participants_dir / file_name
        download_file(participant["profile"], save_path)

    # 게시물 이미지 다운로드
    posts_dir = base_dir / "posts"
    posts_dir.mkdir(exist_ok=True)
    for i, post in enumerate(magazine_data["posts"], start=1):
        file_name = f"post_{i}_{post['member_id']}.jpg"
        save_path = posts_dir / file_name
        download_file(post["image"], save_path)

    return base_dir


def draw_text_with_stretch(c, text, x, y, font_name, font_size, color, stretch_percent):
    # 텍스트 색상 및 폰트 설정
    c.setFillColor(color)
    c.setFont(font_name, font_size)
    
    # stretch_percent에 따른 조정된 가변 간격 계산
    stretch_factor = stretch_percent / 100.0
    
    # 텍스트를 한 글자씩 출력
    for char in text:
        # 각 글자의 실제 너비 계산
        char_width = c.stringWidth(char, font_name, font_size)
        
        # 장평 조정된 간격으로 출력
        c.drawString(x, y, char)
        
        # 가변 간격 적용
        x += char_width * stretch_factor



def create_circular_image(input_image_path, output_image_path):
    # 원본 이미지를 열고 원형으로 자르기
    with Image.open(input_image_path) as im:
        # 정사각형을 만들어 원형으로 자르기 위한 크기 계산
        size = min(im.size)  # 짧은 쪽 기준으로 자르기
        left = (im.width - size) / 2
        top = (im.height - size) / 2
        right = (im.width + size) / 2
        bottom = (im.height + size) / 2
        im = im.crop((left, top, right, bottom))

        # 원형 마스크 생성
        mask = Image.new("L", im.size, 0)
        draw = ImageDraw.Draw(mask)
        draw.ellipse((0, 0, im.size[0], im.size[1]), fill=255)

        # 원형 마스크를 적용하여 투명 배경을 가진 원형 이미지 생성
        circular_im = Image.new("RGBA", im.size)
        circular_im.paste(im, (0, 0), mask=mask)

        # 원형 이미지 저장 (고해상도 유지)
        circular_im.save(output_image_path, format="PNG")

def image_month(url, base_dir):
    # 이미지 다운로드
    image_path = f"{base_dir}/image_month.jpg"
    download_file(url, image_path)
    
    
    with Image.open(image_path) as im:
        # 이미지 크기 계산
        width, height = im.size
        if width > height:
            new_width = height * (152 /110)
            left = (width - new_width) / 2
            right = (width + new_width) / 2
            top, bottom = 0, height
        else:
            new_height = width * (110 /152)
            top = (height - new_height) / 2
            bottom = (height + new_height) / 2
            left, right = 0, width
        
        # 직사각형으로 자르기
        im = im.crop((left, top, right, bottom))
        im.save(image_path, format="PNG")
    
        return image_path
    
def split_and_mask_image(original_image_path):
    from pathlib import Path
    original_image_path = Path(original_image_path)
    
    # 원본 이미지 로드
    with Image.open(original_image_path) as img:
        width, height = img.size
        
        # 전체 이미지에서 필요한 비율만큼 좌우로 분할
        total_ratio = 15 + 143  # 전체 비율 합계
        
        # 왼쪽 이미지 (15:110 비율)
        left_width = int((15 / total_ratio) * width)
        back_image = img.crop((0, 0, left_width, height))
        
        # 오른쪽 이미지 (143:110 비율)
        right_width = int((143 / total_ratio) * width)
        right_x = width - right_width
        front_image = img.crop((right_x, 0, width, height))
        
        # 이미지 저장 (해상도 유지)
        back_image_path = original_image_path.parent / "back_image.png"
        back_image.save(back_image_path, format="PNG")
        
        front_image_path = original_image_path.parent / "front_image.png"
        front_image.save(front_image_path, format="PNG")
    
    return back_image_path, front_image_path


def build_front(c, magazine_data=dict, base_dir=Path, front_image=Path):

    # magazin_data의 uploaded_date 를 리스트를 가지고 온 뒤 가장 최근 날짜를 year, month로 저장
    uploaded_date = []
    for post in magazine_data["posts"]:
        uploaded_date.append(post["upload_date"])
    uploaded_date.sort()
    year = uploaded_date[-1][:4]
    month = uploaded_date[-1][5:7]
    
    # 'year' 텍스트 너비 계산
    c.setFont("NotoSansKR-Light", 20)
    year_text_width = c.stringWidth(year, "NotoSansKR-Light", 20)
    
    # 'month' 텍스트 너비 계산
    c.setFont("NotoSansKR-Bold", 40)
    month_text_width = c.stringWidth(month, "NotoSansKR-Bold", 40)
    
    year_month_x = 18 * mm  # 'year'와 'month'의 시작 X 좌표
    year_month_y = 210 * mm  # 'year'와 'month'의 시작 Y 좌표

    # 'year'의 시작 X 좌표
    year_x = year_month_x
    
    # 'month'의 시작 X 좌표
    month_x = year_month_x - ((month_text_width - year_text_width) / 2)

    # 텍스트 색상 설정 (RGB 값)
    text_color = Color(87/255, 89/255, 91/255)  # #57595B 색상
    black = Color(0, 0, 0)

    c.setFillColor(text_color)

    # 텍스트 출력
    c.setFont("NotoSansKR-Light", 20)
    c.drawString(year_x, year_month_y, year)
    
    c.setFont("NotoSansKR-Bold", 40)
    c.drawString(month_x, year_month_y - 13 * mm, month)

    c.setFont("PrestigeEliteStd-Bd", 12)
    The_Family_length = c.stringWidth("The Family", "PrestigeEliteStd-Bd", 12, )
    c.drawString(170 * mm - The_Family_length, 244 * mm, "The Family")
    Newspaper_length = c.stringWidth("Newspaper", "PrestigeEliteStd-Bd", 12)
    c.drawString(170 * mm - Newspaper_length, 240 * mm, "Newspaper")

    info_x = 70 * mm
    
    draw_text_with_stretch(c, magazine_data["magazine_name"], info_x, 211 * mm, "GowunBold", 16, black, 120)

    draw_text_with_stretch(c, "가정의 달", info_x, 202 * mm, "GowunRegular", 12, text_color, 120)
    draw_text_with_stretch(c, "Authors:", info_x, 192 * mm, "GowunRegular", 12, text_color, 120)

    profile_display=20

    # 시작 위치 및 칼럼 초기 설정
    start_x, start_y = 70 * mm, 180 * mm
    x, y = start_x, start_y
    column_width = profile_display + 26 * mm  # 이미지와 이름 포함한 칼럼 너비
    stretch_percent = 120

    for i, participant in enumerate(magazine_data["participants"], start=1):
        # 원형으로 클리핑된 이미지 생성
        image_path = base_dir / "participants" / f"participant_{i}_{participant['member_id']}.jpg"
        circular_image_path = base_dir / f"circular_participant_{i}.png"
        create_circular_image(image_path, circular_image_path)

        # 원본 크기 유지된 원형 이미지를 PDF에 삽입 시 display_size로 축소
        c.drawImage(str(circular_image_path), x, y, profile_display, profile_display, mask="auto")

        # 이름을 이미지 오른쪽에 장평 조정하여 표시
        name_x = x + profile_display + 2 * mm
        name_y = y + (profile_display / 2) - 4
        draw_text_with_stretch(c, participant["name"], name_x, name_y, "GowunRegular", 10, text_color, stretch_percent)

        # 다음 참가자 위치 조정
        y -= profile_display + 2 * mm

        # 4명씩 한 칼럼을 채운 후 다음 칼럼으로 이동
        if i % 4 == 0:
            x += column_width  # 다음 칼럼으로 이동
            y = start_y  # y 위치를 초기값으로 재설정


    image_x = 5 * mm
    image_y = 30 * mm
    image_width = 152 * mm
    image_height = 110 * mm
    border_color = Color(0xDC/255, 0xDD/255, 0xDD/255)
    border_thickness = 8
    c.drawImage(front_image, image_x, image_y, image_width, image_height)
    c.setStrokeColor(border_color)
    c.setLineWidth(border_thickness)
    
    # 위쪽 테두리
    c.line(5 * mm, image_y + image_height + border_thickness / 2,
           image_x + image_width + border_thickness, image_y + image_height + border_thickness / 2)
    
    # 오른쪽 테두리
    c.line(image_x + image_width + border_thickness / 2, image_y - border_thickness / 2,
           image_x + image_width + border_thickness / 2, image_y + image_height + border_thickness / 2)
    
    # 아래쪽 테두리
    c.line(5 * mm, image_y - border_thickness / 2,
           image_x + image_width + border_thickness, image_y - border_thickness / 2)
    
    c.showPage()

def build_content(c, magazine_data=dict, base_dir=Path):
    # Constants for layout
    margin_left = 13 * mm
    margin_right = 13 * mm
    b5_width, b5_height = 176 * mm, 250 * mm  # B5 크기
    page_width, page_height = b5_width + 10 * mm, b5_height + 10 * mm  # B5 + 5mm 여백 (총 PDF 페이지 크기)
    image_width = page_width - margin_left - margin_right
    image_height = image_width * (350 / 647)
    profile_size = 48  # Profile image size in pixels
    
    # Text color
    text_color = Color(87/255, 89/255, 91/255)  # #57595B
    
    # Process all 10 pages regardless of post count
    posts = magazine_data["posts"]
    
    for page in range(10):
        # Create crop marks for each page
        c = create_b5_canvas_with_outer_edge_crop_marks(c, page_width, page_height)
        
        # Process up to 2 posts per page if available
        for i in range(2):
            post_index = page * 2 + i
            
            # Skip if we've run out of posts
            if post_index >= len(posts):
                break
                
            post = posts[post_index]
            
            # Calculate position based on odd/even index
            if i == 0:  # Top post
                y_position = page_height / 2 + 2.5 * mm
                # 상단 포스트의 프로필 위치 (이미지 위로)
                profile_y = y_position + image_height + 20
                profile_x = margin_left + 16 * mm
            else:  # Bottom post
                y_position = page_height / 2 - 2.5 * mm - image_height
                # 하단 포스트의 프로필 위치 (이미지 아래로)
                profile_y = y_position - profile_size - 20
                profile_x = page_width - margin_right - profile_size - 16 * mm
            
            # Draw the main image
            image_path = base_dir / "posts" / f"post_{post_index + 1}_{post['member_id']}.jpg"
            c.drawImage(str(image_path), margin_left, y_position, image_width, image_height)
            
            # Use the pre-created circular profile image
            circular_profile_path = base_dir / f"circular_participant_{post['member_id']}.png"
            c.drawImage(str(circular_profile_path), profile_x, profile_y, profile_size, profile_size, mask='auto')
            
            # Add uploader name and date with adjusted position based on top/bottom post
            c.setFillColor(text_color)
            name_x = margin_left + profile_size + 5
            date_x = name_x  # 날짜의 x 좌표는 이름과 동일
            date_length = c.stringWidth(post['upload_date'], "GowunRegular", 9)
            name_length = c.stringWidth(post['uploader_name'], "GowunBold", 13)
            c.setLineWidth(0.25)
            c.setStrokeColor(text_color)

            if i == 0:  # Top post
                name_y = profile_y + (profile_size/2) - 4
                name_x = profile_x + profile_size + 5 * mm
                date_y = profile_y + (profile_size/2) - 2
                date_x = page_width - date_length - margin_right - 50
                c.line(name_x + name_length + 7 * mm, name_y + 6 , date_x - 24 * mm, name_y + 6)
            else:  # Bottom post
                name_y = profile_y + (profile_size/2) - 5
                name_x = profile_x - name_length - 7 * mm
                date_y = profile_y + (profile_size/2) - 2
                date_x = margin_left + 50
                c.line(name_x - 7 * mm, name_y + 6 , date_x + date_length + 24 * mm, name_y + 6)
            
            # Add uploader name
            draw_text_with_stretch(c, post['uploader_name'], name_x, name_y, "GowunBold", 13, text_color, 130)
            
            # Add upload date
            draw_text_with_stretch(c, post['upload_date'], date_x, date_y, "GowunRegular", 9, text_color, 120)
        
        # End the page
        c.showPage()
```

```python
def build_back(c, magazine_data=dict, base_dir=Path, back_image=Path):
    b5_width, b5_height = 176 * mm, 250 * mm  # B5 크기
    page_width, page_height = b5_width + 10 * mm, b5_height + 10 * mm
    c = create_b5_canvas_with_outer_edge_crop_marks(c, page_width, page_height)

    image_x = page_width - 8 * mm - 12 * mm
    image_y = 30 * mm
    image_width = 12 * mm + 3 * mm
    image_height = 110 * mm
    border_color = Color(0xDC/255, 0xDD/255, 0xDD/255)
    border_thickness = 8

    c.drawImage(back_image, image_x, image_y, image_width, image_height, anchor='w', mask='auto')
    c.setStrokeColor(border_color)
    c.setLineWidth(border_thickness)
    
    # 위쪽 테두리
    c.line(image_x - border_thickness, image_y + image_height + border_thickness / 2,
           page_width - 5 * mm, image_y + image_height + border_thickness / 2)
    
    # 왼쪽 테두리
    c.line(image_x - border_thickness / 2, image_y - border_thickness / 2,
        image_x - border_thickness / 2, image_y + image_height + border_thickness / 2)
    
    # 아래쪽 테두리
    c.line(image_x - border_thickness, image_y - border_thickness / 2,
           page_width - 5 * mm, image_y - border_thickness / 2)
```

```python
def build_magazine(magazine_data=dict, call_time=str):
    # B5 페이지 크기와 전체 페이지 크기 설정
    b5_width, b5_height = 176 * mm, 250 * mm  # B5 크기
    page_width, page_height = b5_width + 10 * mm, b5_height + 10 * mm  # B5 + 5mm 여백 (총 PDF 페이지 크기)
    fname = f'outputs/{call_time}_magazine{magazine_data["magazine_id"]}.pdf'

    # 폰트 로드
    pdfmetrics.registerFont(TTFont("NotoSansKR-Light", "assets/fonts/NotoSansKR-Light.ttf"))
    pdfmetrics.registerFont(TTFont("NotoSansKR-Bold", "assets/fonts/NotoSansKR-Bold.ttf"))
    pdfmetrics.registerFont(TTFont("GowunBold", "assets/fonts/GowunBatang-Bold.ttf"))
    pdfmetrics.registerFont(TTFont("GowunRegular", "assets/fonts/GowunBatang-Regular.ttf"))
    pdfmetrics.registerFont(TTFont("PrestigeEliteStd-Bd", "assets/fonts/PrestigeEliteStd-Bd.ttf"))

    # Canvas 생성
    c = canvas.Canvas(fname, pagesize=(page_width, page_height))
    # call_time_temp folder 생성
    call_time_temp = f"temp/{call_time}"
    os.makedirs(call_time_temp, exist_ok=True)

    photo_of_month = image_month('https://i.pinimg.com/originals/cd/bc/f0/cdbcf077b62246123f74fcc919587b0b.jpg', f"temp/{call_time}")
    back_image, front_image = split_and_mask_image(photo_of_month)

    base_dir = process_magazine_data(magazine_data, call_time)
    create_b5_canvas_with_outer_edge_crop_marks(c, page_width, page_height)
    build_front(c, magazine_data, base_dir, front_image)
    build_content(c, magazine_data, base_dir)
    build_back(c, magazine_data, base_dir, back_image)

    # PDF 저장
    c.save()
    return fname

build_magazine(magazine_data, "2023-07-11")

#remove temp folder
import shutil
shutil.rmtree("temp")

```
